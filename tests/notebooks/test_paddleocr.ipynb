#### 一、预处理

```python
# third-party
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_img(img_path: str) -> np.ndarray:
    return cv2.imread(img_path)

def preprocess(img: np.ndarray) -> np.ndarray:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # 灰度化
    _, img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)  # 二值化
    return img

def show_img(img: np.ndarray) -> None:
    plt.imshow(img, cmap="gray")
```

```python
def main() -> None:
    img_path = r"C:/UserData/Projects/uv_project/paddleocr_tbd/figs/yyslz_djq.png"
    show_img(preprocess(read_img(img_path)))


main()
```
